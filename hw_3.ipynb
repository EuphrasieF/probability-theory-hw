```python
import math
import matplotlib.pyplot as plt
from scipy import special
from scipy.special import factorial
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
def f_binomial(p, N, n):
    return (special.binom(N, n) * p ** n * (1 - p) ** (N - n))
```

```python
def f_poisson(p, N, n):
    return ((N * p) ** n * math.exp(- N * p) / factorial(n))
```

```python
def f_moivre_laplace(p, N, n):
    sigma_squared = p * (1 - p)
    delta_x = 1 / math.sqrt(N)
    return (delta_x * math.exp(- (n - N * p) ** 2 / (2 * sigma_squared * N)) / math.sqrt(2 * sigma_squared * math.pi))
```

```python
f_binomial_arr = []
f_poisson_arr = []
f_moivre_laplace_arr = []
n_arr = []
```

```python
p = float(input()) #от 0 до 1
N = int(input()) #натуральное число
```

```python
for n in range(30):
    n_arr.append(n)
    f_binomial_arr.append(f_binomial(p, N, n))
    f_poisson_arr.append(f_poisson(p, N, n))
    f_moivre_laplace_arr.append(f_moivre_laplace(p, N, n))


plt.scatter(n_arr, f_binomial_arr, c = 'b') #синие точки - биномиальное распределение
plt.scatter(n_arr, f_poisson_arr, c = 'g') #зеленые точки - распределение Пуассона
plt.scatter(n_arr, f_moivre_laplace_arr, c = 'r') #красные точки - асимптотическое приближение биномиального распределения
plt.show()
```
