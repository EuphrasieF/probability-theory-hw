```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
a = float(input())
b = float(input())
f_arr = []
```

```python
for i in range(100000):
    n = np.random.uniform(0, 1)
    e = np.random.uniform(0, 1)
    f_arr.append(n ** a + e ** b)

n, bins, patches = plt.hist(f_arr, 50, alpha=0.75)
plt.show()
```
