```python
import matplotlib.pyplot as plt
import random as rnd
import math
```

```python
%matplotlib inline
```

```python
x_arr = []
f_arr = []

for i in range(1000):
    x_1 = rnd.uniform(0, 1)
    x_2 = rnd.uniform(0, 1)
    x_arr.append(math.sqrt(-2*math.log(x_1))*math.sin(2*math.pi*x_2))
    f_arr.append(1/(math.sqrt(2 * math.pi)) * math.exp(- x_arr[i]*x_arr[i]/2))
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
plt.scatter(x_arr, f_arr) #рисуем нормальное распределение
plt.show()
```

Как и следовало ожидать, максимум функции соответствует значению $$f = \frac{1}{\sqrt{2\pi}}$$
