```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
def diy_normal_distribution():
    xi = np.random.uniform(0, 1, size=2)
    return np.sqrt(-2 * np.log(xi[0])) * np.sin(2 * np.pi * xi[1])
```

```python
def normal_distribution_density(x):
    return 1 / (np.sqrt(2 * np.pi)) * np.exp(-x ** 2 / 2)
```

```python
def build_hist_with_density():
    x_values = np.linspace(-10, 10, 500)
    y_values = np.vectorize(normal_distribution_density)(x_values)
    plt.plot(x_values, y_values, color='red')

    samples = np.fromiter((diy_normal_distribution() for _ in range(5000)), float)
    plt.hist(samples, bins=50, alpha=0.5, density=True)

    plt.show()
```

```python
build_hist_with_density()
```

Как и следовало ожидать, максимум функции соответствует значению $$f = \frac{1}{\sqrt{2\pi}}$$
