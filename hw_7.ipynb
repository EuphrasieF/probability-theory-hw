```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
def symmetric_random_matrix(N):
    A = np.random.normal(size=(N, N))
    U = np.triu(A)
    U_ = np.triu(A, k=1)
    return U + U_.T
```

```python
symmetric_random_matrix(4)
```

```python
def xi(A):
    eigenvalues, _eigenvectors = np.linalg.eig(A)
    return eigenvalues / np.sqrt(np.shape(A)[0])
```

```python
A = symmetric_random_matrix(4)
```

```python
xi(A)
```

```python
def wigner_distribution_density(x):
    return (1 / (2 * np.pi)) * np.sqrt(4 - x ** 2)
```

```python
def build_hist_with_density(N):
    x_values = np.linspace(-2.0, 2.0, 100)
    y_values = np.vectorize(wigner_distribution_density)(x_values)
    plt.plot(x_values, y_values, color='red')

    samples = xi(symmetric_random_matrix(N))
    plt.hist(samples, bins=10, alpha=0.5, density=True)

    plt.show()
```

```python
build_hist_with_density(200)
```
