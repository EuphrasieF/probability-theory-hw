```python
import matplotlib
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
def diy_exp_distribution(lambda_):
    xi = np.random.uniform(0, 1)
    return -(1/lambda_) * np.log(xi)
```

```python
def exp_distribution_density(lambda_, x):
    return lambda_ * np.exp(-lambda_ * x)
```

```python
def build_hist_with_density(lambda_):
    x_values = np.linspace(1e-3, 10, 100)
    y_values = np.vectorize(lambda x: exp_distribution_density(lambda_, x))(x_values)
    plt.plot(x_values, y_values, color='red')

    samples = np.fromiter((diy_exp_distribution(lambda_) for _ in range(5000)), float)
    plt.hist(samples, bins=50, alpha=0.5, density=True)

    plt.show()
```

```python
build_hist_with_density(1)
```

```python
build_hist_with_density(0.5)
```

```python
build_hist_with_density(2.5)
```
