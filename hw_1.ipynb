```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
l = float(input())
```

```python
e_arr = [] #массив эта
n_arr = [] #массив кси

for i in range(1000):
    y = np.random.uniform(0, 1) #рандомное кси
    n_arr.append(y) #добавляем кси в массив
    e_arr.append(-math.log(n_arr[i])/l) #вычисляем эта

plt.scatter(n_arr, e_arr, marker=",") #строим графит кси от эта
plt.show()
```

```python
l_1 = 1
l_2 = 10
l_3 = -10
```

```python
#сделаем на одной картинке три графика для разных лямбда

e_1_arr = [] 
n_1_arr = [] 

e_2_arr = [] 
n_2_arr = [] 

e_3_arr = [] 
n_3_arr = [] 

for i in range(1000):
    y = np.random.uniform(0, 1) 
    n_1_arr.append(y) 
    e_1_arr.append(-math.log(n_1_arr[i])/l_1) 
    
for i in range(1000):
    y = np.random.uniform(0, 1) 
    n_2_arr.append(y) 
    e_2_arr.append(-math.log(n_2_arr[i])/l_2)
    
for i in range(1000):
    y = np.random.uniform(0, 1) 
    n_3_arr.append(y) 
    e_3_arr.append(-math.log(n_3_arr[i])/l_3) 
    
plt.rcParams['figure.figsize'] = [15, 10]
    
plt.scatter(n_1_arr, e_1_arr, marker = ",")
plt.scatter(n_2_arr, e_2_arr)
plt.scatter(n_3_arr, e_3_arr)

plt.show()
```
